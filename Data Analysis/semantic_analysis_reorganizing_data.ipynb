# 3 semantics

## Datasets

```python
ds_path = "/Datasets/analysis/Reddit_SNS_semantic_3.csv" 
df = pd.read_csv(ds_path, encoding='latin-1')
df
```

```python
labeled_df = []
for item in range(len(df)):
  scores = [df['neg'][item], df['neu'][item], df['pos'][item]]
  max_score = scores.index(max(scores))
  term = [df['Text'][item], max_score, df['compound'][item]]
  labeled_df.append(term)

labeled_df = pd.DataFrame(labeled_df, columns=['Text', 'Score', 'Compound'])
labeled_df
```

```python
save_path = "/Datasets/analysis/"
labeled_df.to_csv(save_path+'Reddit_SNS_semantic_3_refactored.csv')
```

# 6 Semantics

```python
ds_path = "/Datasets/analysis/Reddit_SNS_semantic_6.csv" 
df = pd.read_csv(ds_path, encoding='latin-1')
df
```

```python
labeled_df = []
for item in range(len(df)):
  scores = [df['sadness'][item], df['joy'][item], df['love'][item], df['anger'][item], df['fear'][item], df['surprise'][item]]
  max_score = scores.index(max(scores))
  term = [df['Text'][item], max_score]
  labeled_df.append(term)

labeled_df = pd.DataFrame(labeled_df, columns=['Text', 'Score'])
labeled_df
```

```python
save_path = "/Datasets/analysis/"
labeled_df.to_csv(save_path+'Reddit_SNS_semantic_6_refactored.csv')
```
